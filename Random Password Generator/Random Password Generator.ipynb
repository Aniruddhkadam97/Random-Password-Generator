```python
#importing
import tkinter
import tkinter.font as font
from time import strftime

import webbrowser
import string
import random
from random import randint
import tkinter
import tkinter as tk
import tkinter.font as TkFont
from tkinter import *
from tkinter import ttk
from tkinter.messagebox import *
from tkinter.simpledialog import askstring
from tkinter import messagebox

import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2
import pandas as pd
import numpy as np

import numpy
from openpyxl import *
import secrets





root = tk.Tk()
style = ttk.Style()

root.geometry("1000x525")
root['bg'] = "HoneyDew"
root.resizable(False, False)

root.title('Random Password Generator')

####################################################################################################

def Random_Password_Generator():
    wb = load_workbook('data_excel.xlsx') # 

    sheet = wb.active # permission to get activated like read write 

    def excel():
        print("excel")


    def clear():
        name_field.delete(0, END)
        Pass_len_field.delete(0, END)
        Password_field.delete(0, END)

    def Gen_pas():
        Password_field.delete(0, END)
        len=Pass_len_field.get()
        import ps_word
        if Pass_len_field.get() == "":
            messagebox.showwarning("Warning!", "Empty Password Length")

        elif  int(len) >= 6 and int(len) <= 64:
           s_pass=ps_word.password_dt(len)
           Password_field.insert(0,s_pass)

        else:
            messagebox.showwarning("Warning!", "Password Not between 6 and 64 Characters")


    def insert():
        if name_field.get() == "" and Pass_len_field.get() == "" and Password_field.get() == "":
            messagebox.showwarning("Warning!", "Fields cannot be empty")
        elif (name_field.get().isalpha()==False): 
            messagebox.showwarning("Warning!", "Name Must be only String")


        else:

            current_row = sheet.max_row# anydata from text field is atored here 
            current_column = sheet.max_column

            sheet.cell(row=current_row + 1, column=1).value = name_field.get()
            sheet.cell(row=current_row + 1, column=2).value = Pass_len_field.get()
            sheet.cell(row=current_row + 1, column=3).value = Password_field.get()

            wb.save('data_excel.xlsx')

            name_field.focus_set()

            clear()


    if __name__ == "__main__":
        root = Tk()
        excel()

        root.geometry("700x490")
        root.maxsize(width="700",height="490")
        root.minsize(width="700",height="490")
        root.config(bg="deep sky blue")
        root.title("Random Password Generator")

        signup_frame = Frame(root,width="500",height="400",bg="seashell")
        signup_frame.place(x="65",y="140")

        heading = Label(root,text="Random Password Generator",font=("Arial",36,"bold"),padx="5",pady="5",bg="deep sky blue",fg="HoneyDew", borderwidth=3,  relief="solid").pack(pady="50")

        name = Label(
            signup_frame,
            width="20",
            padx="5",
            pady="5",
            text="Name",
            bg="lavender",
            fg="darkgreen",
            font=("Calibri",15,'bold'),
            borderwidth=1,
            relief="solid"
        ).grid(
            row=0,
            column=0,
            padx="10",
            pady="10"
        )

        sem = Label(
            signup_frame,
            width="20",
            padx="5",
            pady="5",
            text="Generated Password",
            bg="lavender",
            fg="darkgreen",
            font=("Calibri",15,'bold'),
            borderwidth=1,
            relief="solid"
        ).grid(
            row=2,
            column=0,
            padx="10",
            pady="10"
        )

        course = Label(
            signup_frame,
            width="20",
            padx="5",
            pady="5",
            text="Password Length",
            bg="lavender",
            fg="darkgreen",
            font=("Calibri",15,'bold'),
            borderwidth=1,
            relief="solid"
        ).grid(
            row=1,
            column=0,
            padx="10",
            pady="10"
        )

        name_field = Entry(
            signup_frame,
            width="30",
            selectbackground="azure",
            selectforeground="red",
            font=("Calibri",15,"bold"),
            borderwidth=1,
            relief="solid"
        )
        name_field.grid(
            row=0,
            column=1,
            padx="15",
            pady="5"
        )

        Pass_len_field = Entry(
            signup_frame,
            width="30",
            selectbackground="azure",
            selectforeground="red",
            font=("Calibri",15,"bold"),
            borderwidth=1,
            relief="solid"
        )
        Pass_len_field.grid(
            row=1,
            column=1,
            padx="15",
            pady="5"
        )

        Password_field = Entry(
            signup_frame,
            width="30",
            selectbackground="azure",
            selectforeground="red",
            font=("Calibri",15,"bold"),
            borderwidth=1,
            relief="solid"
        )
        Password_field.grid(
            row=2,
            column=1,
            padx="15",
            pady="5"
        )

        submit = Button(
            signup_frame,
            text="Generate Password",
            width="20",
            bg="mistyrose",
            fg="black",
            font=("Calibri",15,"bold"),
            relief="solid",
            borderwidth=2,
            command=Gen_pas
        ).grid(
            row=3,
            column=1,
            padx="15",
            pady="5"
        )

        submit = Button(
            signup_frame,
            text="Save",
            width="20",
            bg="mistyrose",
            fg="black",
            font=("Calibri",15,"bold"),
            relief="solid",
            borderwidth=2,
            command=insert
        ).grid(
            row=4,
            column=1,
            padx="15",
            pady="5"
        )

        submit = Button(
            signup_frame,
            text="Reset",
            width="20",
            bg="mistyrose",
            fg="black",
            font=("Calibri",15,"bold"),
            relief="solid",
            borderwidth=2,
            command=clear
        ).grid(
            row=5,
            column=1,
            padx="15",
            pady="5"
        )

    file_df = pd.read_excel("data_excel.xlsx")
    file_df.sort_values("User Name", inplace = True)
    bool_series = file_df["User Name"].duplicated()
    file_df.head()#
    data=file_df[bool_series]


    df = pd.DataFrame(data)

    name = df['User Name'].head()
    passw = df['Password Length'].head()


    fig = plt.figure(figsize =(5, 4))


    plt.bar(name[0:10], passw[0:10])

    plt.xlabel("User Name")
    plt.ylabel("Length Of Password")
    plt.title("Password Generator")
    plt.show()
    
####################################################################################################


####################################################################################################

label1 = Label(
    root,
    text = " Random Password Generator ", 
    font = ("Comic Sans MS",50,"bold italic"), 
    borderwidth=3, 
    relief="solid"
)
label1.config(#Times New Roman
    bg= "Ivory", 
    fg= "black"
)
label1.pack(
    pady = 15, 
    padx = 15,
)

button_font = font.Font(family='Georgia', size=20)
 
button_submit1 = tkinter.Button(root,
    text="Random Password Generator",
    bg='#45b592',
    fg='#ffffff',
    bd=2,
    font=button_font,
    height=1,
    width=25,
    command=Random_Password_Generator
).pack(
    ipady = 10, 
    ipadx = 10,
    pady = 10
)

def my_time():
    time_string = strftime('%H:%M:%S %p - %A -  %x') # time format 
    l1.config(text=time_string)
    l1.after(1000,my_time) # time delay of 1000 milliseconds 

my_font=('Georgia',15,'bold') # display size and style

l1=tk.Label(
    root,
    font=my_font,
    bg='HoneyDew'
)
l1.pack(
    padx=5,
    pady=5
)

my_time()

root.mainloop()
```
